##Punto 1

Probar que el vector mas pequeño de tal forma que $2x + 3y =5$ es $u=\begin{pmatrix} 0.76923 \\
1.15385 \end{pmatrix}$

##Solucion

Expresando esto el sistema en terminos matriciales $Ax=b$

$$\begin{pmatrix}
2 & 3
\end{pmatrix}
\begin{pmatrix}
x\\
y
\end{pmatrix}=5$$

vemos que el sistema está subdeterminado, la solucion no es unica ya que $N(A)≠0$ por lo que debemos encontrar el vector $x$ mas pequeño de todas las soluciones $min\|x\|$

Podemos encontrar la solucion general expresando x en terminos de y como

$$x=\frac{5-3y}{2}$$

Por lo tanto el vector x se reescribe como

$$\begin{pmatrix}
\frac{5-3y}{2} \\
y
\end{pmatrix}$$

Osea que todos los vectores $x_n$ que cumplan este criterio forman el espacio solucion de de el sistema.

Para encontrar la solucion con la magnitud mas pequeña, debemos encontrar el minimo de la expresion que calcula la magnitud de los vectores

$$\|x\|=\sqrt{x^{2}+y^{2}}$$

$$\|x\|=\sqrt{(\frac{5-3y}{2})^{2}+y^{2}}$$

podemos encontrar el vector con la longitud mas pequeña, el cual va a ser perpendicular al espacio de solucion y la respuesta del problema de minimos cuadrados (segun la figura vista en clase) derivamos esta expresion respecto a y utilizando la regla de la cadena

$$min\|x\|= \frac{d\|x\|}{dy}=\frac{1}{2}\left((\frac{5-3y}{2})^{2}+y^{2}\right)^{-1/2} \times \frac{d}{dy}\left((\frac{5-3y}{2})^{2}+y^{2}\right)$$
$$\frac{d\|x\|}{dy}=\frac{1}{2}\left((\frac{5-3y}{2})^{2}+y^{2}\right)^{-1/2} \times \left(2(\frac{5-3y}{2})\frac{-3}{2}+2y\right)$$
$$\frac{d\|x\|}{dy}=\frac{1}{2}\left((\frac{5-3y}{2})^{2}+y^{2}\right)^{-1/2} \times \left(\frac{13y-15}{2}\right)$$

Racionalizando y desarrollando la expresion se llega a que

$$min\|x\|= \frac{d\|x\|}{dy}=\frac{13y-15}{2\sqrt{(\frac{5-3y}{2})^{2}+y^{2}}}$$

Igualando esta expresion a 0 se encuentra el y minimo, para obtener el y donde la expresion se hace 0 se utiliza la calculadora grafica desmos
$$\frac{13y-15}{2\sqrt{(\frac{5-3y}{2})^{2}+y^{2}}}=0$$
$$y=\frac{15}{13}≈1.15385$$

y sabiendo que todos los $x$ son de la forma $x=\frac{5-3y}{2}$ se calcula la solucion de $x$ con el valor obtenido anteriormente de $y$ como
$$x=\frac{10}{13}≈0.76923$$

Por lo tanto el vector $u$ mas pequeño, la cual es la solucion de minimos cuadrados del sistema es

$$u=\begin{pmatrix}
0.76923 \\
1.15385
\end{pmatrix}$$

_Se termina la prueba_

##Punto 2

Encuentra la pseudoinversa de la matriz $A=\begin{pmatrix} 2 & 3 \end{pmatrix}$
Pista: Recuerde que la pseudoinversa $A^{+}$ debe satisfacer $A^{+}b=u$. $b$ puede ser cualquier numero $\in \mathbb{R} >0$. No usar SVD, solo resolver ecuaciones lineales

##Solucion

Sabiendo que la pseudoinversa debe satisfacer $A^{+}b=u$, siendo u el vector con longitud mas pequeña, perpendicular al espacio solucion, y sabiendo que estamos trabajando con la misma matriz, podemos generalizar la forma de encontrar u para cualquier b siguiendo el mismo procedimiento realizado en el punto anterior, la expresion para $min\|x\|$ resulta como

$$\frac{d\|x\|}{dy} = \frac{13y - 3b}{2\sqrt{\left(\frac{b - 3y}{2}\right)^2 + y^2}}$$

Igualando esta expresion a 0 y utilizando gemini para simplificar el polinomio.

$$\frac{13y - 3b}{2\sqrt{\left(\frac{b - 3y}{2}\right)^2 + y^2}}=0$$

encontramos que los $x$ y $y$ minimos para cualquier $b$ son de la forma

$$x=\frac{2b}{13} \quad y=\frac{3b}{13}$$

por lo tanto el vector $u$ para cualquier $b$ se define como

$$u=\begin{pmatrix} \frac{2b}{13} \\ \frac{3b}{13} \end{pmatrix}$$

Sabiendo que $A^{+}b=u$ tenemos que

$$A^{+}b=\begin{pmatrix} \frac{2b}{13} \\ \frac{3b}{13} \end{pmatrix}$$

Sabiendo que $b\in\mathbb{R}>0$ podemos dividir a ambos lados por $b$ y obtenemos
$$A^{+}=\begin{pmatrix} \frac{2}{13} \\ \frac{3}{13} \end{pmatrix}$$

Esta seria la Pseudoinversa $A^{+}$ de la matriz $A$, para comprobar esto, podemos calcular $A^{+}b$ con $b=5$ como en el punto anterior.

$$A^{+}b=\begin{pmatrix} \frac{2}{13} \\
\frac{3}{13}
\end{pmatrix}
\times 5
=\begin{pmatrix} \frac{10}{13} \\
\frac{15}{13} \end{pmatrix} \approx\begin{pmatrix} 0.76923 \\
1.15385 \end{pmatrix}$$

Ya La solucion de $A^{+}b$ con la Matriz $A{+}$ calculada y $b=5$ es el vector $u$ calculado en el punto anterior, se cumple que $A^{+}b=u$ por lo que se confirma que la matriz $A^{+}$ Calculada es la Pseudoinversa de la matriz $A$

_Se termina la prueba_
