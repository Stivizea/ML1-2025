## Punto 1
Pruebe que la suma de matrices as asociativa.

*Solucion*

Definimos tres matrices $A$, $B$, $C$ de tal forma que $A$,$B$,$C \in \mathbb{C}^{m \times n} $

Utilizando compontentes

$$ A=(a_{ij}),\quad B=(b_{ij}), \quad C=(c_{ij}) \quad i=1,...,m\quad j=1,...,n$$

 $A+(B+C)$ por definicion de suma de matrices se puede escribir como

$$ A+(B+C)= (a_{ij}) + ((b_{ij})+(c_{ij}))$$

y $(A+B)+C$ como

$$ (A+B)+C= ((a_{ij}) + (b_{ij}))+(c_{ij})$$

ya que $(a_{ij}),(b_{ij}),(c_{ij}) \in \mathbb{C}$, debido a la propiedad de asociatividad de los numeros complejos, estas expresiones se pueden igualar como:

$$  (a_{ij}) + ((b_{ij})+(j_{ij}))=((a_{ij}) + (b_{ij}))+(c_{ij})$$

Y, debido a la definicion de suma de matrices, esta expresion es equivalente a

$$A+(B+C) = (A+B)+C$$

Se finaliza la prueba

##PUNTO 2.1##
Probar que $(AB)^T=B^TA^T$

_Solucion_

Se definen las matrices $A,B,C,D \in \mathbb{c}^{n\times m}$
de tal forma que

$$ A=(a_{ij}),\quad B=(b_{ij}),\quad C=(c_{ij}) \quad D=(d_{ij})   \quad i=1,...,m\quad j=1,...,n$$

Segun la definicion de producto de matrices se define C como:

$$C=AB=(\sum_{k=1}^n a_{ik} b_{kj})=(c_{ij})$$

Segun la definicion de matriz transpuesta y de producto de matrices

$$C^T=(c_{ji})=(\sum_{k=1}^n a_{jk} b_{ki})=(AB)^T $$

Segun la definicion de Matriz transpuesta, se define D como

$$D=B^TA^T=(\sum_{k=1}^n (b_{ik})^T (a_{kj})^T)=(\sum_{k=1}^n b_{ki} a_{jk})$$

ya que $a_{ik}$ y $b_{ki}\in\mathbb{C}$, por la propiedad comutativa de los numeros complejos

$$(\sum_{k=1}^n a_{jk} b_{ki})=(\sum_{k=1}^n b_{ki} a_{jk})$$

y reemplazando las definiciones de estas sumatorias que se hicieron anteriormente

$$C^T=D$$
por lo tanto
$$(AB)^T =B^TA^T$$

prueba terminada

##Punto 2.2
Probar que $(A^T)^T=A$

_Solucion_

Definimos las matrices $A,C\in\mathbb{C}^{m\times n}$, utilizando componentes

$$ A=(a_{ij}),\quad C=(c_{ij}) \quad i=1,...,m\quad j=1,...,n$$

Por la definicion de matriz transpuesta

$$ A^T=(a_{ji})$$
Definimos la matriz $C$ como
$$C=(c_{ij})=A^T=(a_{ji})$$

Por tanto y por la definicion de matriz transpuesta, $C^T$ se puede definir como

 $$C^T=(c_{ji})=(a_{ij})=A$$
desarrollando esta expresion
$$C^T=A$$
Reemplazando segun la definicion de la matriz $C$ hecha anteriormente
$$(A^T)^T=A$$

Se termina la prueba

##Punto 3
Probar que el producto Hadamard de matrices es conmutativo

_Solucion_

Definimos dos matrices $A$, $B$,  de tal forma que $A$,$B \in \mathbb{C}^{m \times n} $

Utilizando compontentes

$$ A=(a_{ij}),\quad B=(b_{ij}), \quad i=1,...,m\quad j=1,...,n$$

Segun la definicion de producto Hadamard
$$A\circ B =(a_{ij}b_{ij})$$

$$B\circ A =(b_{ij}a_{ij})$$

Ya que $(a_{ij}),(b_{ij})\in\mathbb{C}$, segun la propiedad conmutativa de los numeros complejos:

$$(a_{ij}b_{ij})=(b_{ij}a_{ij})$$

Utilizando la definicion de producto Hadamard, esta expresion es equivalente a

$$A\circ B = B\circ A$$

Por lo que el producto Hadamard de matrices es conmutativo, se termina la prueba

Muestre un ejemplo donde el producto de matrices no es conmutativo (Asuma que las matrices son cuadradas)

Definamos dos matrices cuadradas $A,B\in\mathbb{R}^{2 \times 2}$

$$A = \begin{pmatrix}
2 & 3 \\ -3 & 2 \end{pmatrix}, \quad B = \begin{pmatrix}
1 & 8 \\ 4 & -6 \end{pmatrix}
$$

Utilizando la definicion de producto de matrices

$$AB= (\sum_{k=1}^2 a_{ik} b_{kj})= \begin{pmatrix}
2*1+3*4 & 2*8-3*6 \\ -3*1+2*4 & -3*8-2*6 \end{pmatrix}=\begin{pmatrix}
14 & -2\\ 5 & -36 \end{pmatrix}$$

Esto se puede comprobar en el siguiente codigo


```python
import numpy as np

A = np.array( [[2,3], [-3,2]])
B = np.array( [[1,8], [4,-6]])
C = A.dot(B)
print(C)
```

Utilizando la definicion de suma de matrices

$$BA= (\sum_{k=1}^2 b_{ik} a_{kj})= \begin{pmatrix}
1*2-8*3 & 1*3+8*2\\ 4*2+6*3 & 4*3-6*2 \end{pmatrix}=\begin{pmatrix}
-22 & 19\\ 26 & 0 \end{pmatrix}$$

Esto se puede comprobar en el siguiente codigo

```python
import numpy as np

A = np.array( [[2,3], [-3,2]])
B = np.array( [[1,8], [4,-6]])
C = B.dot(A)
print(C)
```

Por tanto se comprueba que para este ejemplo  $AB \neq BA$, en otras palabras, el producto de estas matrices no es conmutativo

#Punto 5
Muestre un ejemplo donde el producto interno no sea conmutativo

_Solucion_

Aunque el producto interno siempre es conmutativo para vectores en $\mathbb{R}$, no siempre lo es para vectores en $\mathbb{C}$, definamos dos vectores $u,v\in\mathbb{C}^{2}$

$$u = \begin{pmatrix} 2-4i & 1+2i\end{pmatrix}\quad v=\begin{pmatrix}7 & -2+3i\end{pmatrix}$$

Segun la definición de producto interno para complejos, se define $\langle u,v \rangle$ como

$$ \langle u, v⟩=\sum_{i=1}^2 u_{i}\overline{v_{i}}=(2-4i)(7)+(1+2i)(-2-3i)=14-28i-2-3i-4i+6=18-35i$$

Cambiando el orden de los vectores en el producto interno
$$ \langle v, u⟩=\sum_{i=1}^2 v_{i}\overline{u_{i}}=(7)(2+4i)+(-2+3i)(1-2i)=14+28i-2+4i+3i+6=18+35i$$

Por tanto, en este ejemplo se puede ver que $⟨u,v⟩\neq ⟨v,u⟩$ y por lo tanto, para estos vectores en particular el producto interno no es conmutativo, lo que se puede confirmar en el siguiente codigo



```python
# verficacion de product interno
z1 = np.array( [2-4j, 1 + 2j])
z2 = np.array( [7,-2+3j])

# inner product of z1 and z2
print(np.vdot(z2,z1))
```
